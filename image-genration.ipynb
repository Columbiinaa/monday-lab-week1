```python
import os
import torch
from diffusers import StableDiffusionPipeline

model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float32
)

pipe = pipe.to("cpu")

output_dir = "synthetic_dataset"
os.makedirs(output_dir, exist_ok=True)

prompts = [
    "A futuristic city at sunset",
    "A cute cat wearing a space helmet",
    "A robot painting a portrait",
    "A peaceful village in the mountains",
    "A cyberpunk street at night"
]

for i, prompt in enumerate(prompts):
    image = pipe(prompt).images[0]
    image.save(f"{output_dir}/image_{i+1}.png")

print("Dataset generated successfully!")
```

```python
import os
import torch
from diffusers import StableDiffusionPipeline

# Load Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float32
)

pipe = pipe.to("cpu")

dog_breeds = [
    "Labrador Retriever",
    "German Shepherd",
    "Golden Retriever",
    "Bulldog",
    "Poodle",
    "Beagle",
    "Rottweiler",
    "Siberian Husky"
]

base_dir = "dog_dataset"
os.makedirs(base_dir, exist_ok=True)

images_per_breed = 2  # you can increase to 5+

for breed in dog_breeds:
    breed_folder = os.path.join(base_dir, breed.replace(" ", "_"))
    os.makedirs(breed_folder, exist_ok=True)

    for i in range(images_per_breed):
        prompt = f"A high quality photo of a {breed} dog"
        image = pipe(prompt).images[0]
        image.save(f"{breed_folder}/img{i+1}.png")

print("Dog breed dataset generated successfully!")
```
